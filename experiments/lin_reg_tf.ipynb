```python
import numpy as np
import tensorflow as tf

import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

%matplotlib inline
```

```python
n=100

x = tf.random.normal(mean=0, stddev=1, shape=(n, 1))
ones = tf.ones(shape=(n, 1))

X = tf.concat([ones, x], axis=1)
```

```python
beta = tf.constant([[1.0], [3.0]])
```

```python
epsilon = tf.random.normal(mean=0, stddev=0.1, shape=(n, 1))

y = tf.linalg.matmul(X, beta) + epsilon 
```

```python
plt.rcParams['figure.figsize'] = (10,7)

fig, ax = plt.subplots()

sns.scatterplot(
    x=X[:, 1].numpy().flatten(), 
    y=y.numpy().flatten(), 
    ax=ax
)

ax.set(title='Raw Data');
```
